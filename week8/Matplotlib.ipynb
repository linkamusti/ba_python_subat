```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5,6]
y = [6,8,19, 23, 4, 1]

plt.plot(x,y) # Line Chart çiziyor

plt.title("Bilge Adam Plotu") # Plot'a başlık ekler.

# Eksenlerin başlıklarını ekler
plt.xlabel("X değerlerim") 
plt.ylabel("Y değerlerim")

plt.show() # çizilen plotu görebilmek için bunu yazmak gerekiyor
```

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5,6]
y = [6,8,19, 23, 4, 1]
y2 = [1,9, 15, 40, -5, 4]
y3 = [1,2,3, 4, 5, 6]

plt.plot(x,y, label = "Birinci", color="red", linestyle="dotted", linewidth=3) # Line Chart çiziyor
plt.plot(x,y2, label = "İkinci", color="green", linestyle="dashed", linewidth=2) # Line Chart çiziyor
plt.plot(x,y3, label = "Üçüncü", color = "blue", linewidth=0.5) # Line Chart çiziyor

plt.title("Bilge Adam Plotu") # Plot'a başlık ekler.

# Eksenlerin başlıklarını ekler
plt.xlabel("X değerlerim") 
plt.ylabel("Y değerlerim")

plt.legend(loc="upper left")

plt.show() # çizilen plotu görebilmek için bunu yazmak gerekiyor
```

```python
import numpy as np

x = np.linspace(0,10, 11)
y = 2 * x
plt.plot(x,y, label = "f(x) = 2x")
plt.plot(x, x**2, label = "f(x) = x**2")
plt.plot(x, x * 4 + 7, label = "f(x) = 4x + 7")


plt.legend()
plt.show()
```

```python
import numpy as np

x = np.linspace(1,30,60)
y = np.sin(x)
y2 = np.cos(x)
y3 = np.log(x)

plt.plot(x,y, label = "sin(x)")
plt.plot(x,y2, label = "cos(x)")
plt.plot(x,y3, label = "log(x)")

plt.fill_between(x, y, y2, alpha=0.1) # İki graph arasını boyamak için

plt.legend()
plt.show()
```

```python
import numpy as np

x = np.arange(0, 5, 0.1)
y = np.exp(-x) * np.cos(2*np.pi*x)

plt.plot(x,y,label="enteresan fonksiyon")
plt.text(4.4, 0.8, "pi=3.14") # plot üzerine ekstra bilgi eklemek için
plt.annotate('burası süper', xy=(1,0.4), xytext=(2, 0.7), arrowprops=dict(shrink=0.01))

plt.legend()
plt.show()
```

```python
## Scatter Plot

# Yöntem 1
# marker list: https://matplotlib.org/stable/api/markers_api.html

x = np.array([1,2,3,4])
y = np.array([5,2,9,3])

plt.plot(x,y,'3')
plt.show()
```

```python
## Scatter Plot

# Yöntem 2
# marker list: https://matplotlib.org/stable/api/markers_api.html

x = np.array([1,2,3,4])
y = np.array([5,2,9,3])
sizes = np.array([50, 100, 150, 200])
colors = np.array([0.1, 0.2, 0.3, 0.4])

plt.scatter(x,y, s=sizes, c=colors, marker="+") # s size'ı belirtiyor, c color'ı belirtiyor
plt.show()
```

```python
# Bar Plots

x = ['C', 'C++', 'Java', 'Python', 'PHP']
y = [23, 17, 35, 29, 12]

plt.bar(x,y)
plt.show()
```

```python
# Yanyana birden fazla bar chart
x = np.array([2017, 2018, 2019, 2020, 2021])
y = np.array([100, 90, 80, 70, 60])
y2 = np.array([60, 70, 80, 90, 100])

plt.bar(x, y, width = 0.4, label = "Yemek")
plt.bar(x + 0.4, y2, width= 0.4, label = "Nüfus")

plt.legend()

plt.show()
```

```python
data = np.random.randint(10, 100, (3,5))
print(data)

x = np.arange(5)

plt.bar(x - 0.25, data[0], width = 0.25, label = "C")
plt.bar(x ,  data[1], width = 0.25, label = "C++")
plt.bar(x + 0.25, data[2], width = 0.25, label = "C#")
plt.legend()
plt.show()
```

```python
#Stacked bar chart

men = np.array([20, 35, 30, 35, 27])
women = np.array([25, 32, 34, 20, 25])
x = np.arange(5)

plt.bar(x, men, width=0.35, color = "r", label = "men")
plt.bar(x, women, width=0.35, bottom=men, color= "b", label="women")

plt.legend()
plt.show()
```

```python
data = [20, 20, 55, 65]

labels = ["C", "C++", "Java", "Python"]
explode = [0,0,0,0.1]
plt.pie(data, labels = labels, autopct='%1.1f%%', 
        shadow=True, startangle=90 ,explode=explode)
plt.show()
```

```python
## Multiple plots in same figure

x = np.arange(5)

fig, ax = plt.subplots(2)

ax[0].plot(x, x*2)
ax[1].plot(x, x**2)

fig.show()
```

```python
## Multiple plots in same figure

x = np.arange(5)

fig, ax = plt.subplots(2,3, gridspec_kw={'hspace': 0.9, 'wspace': 0.5}) # gridspec_kw, subplotlar arasında boşluk bırakmak için var 
                                                                        # hspace dikey, wspace yatay boşluk


ax[0, 0].plot(x, x*2)
ax[0, 0].set_title("0,0'daki plot")


ax[0, 1].plot(x, x**2)
ax[0, 1].set_title("0,1'deki plot")


ax[0, 2].plot(x, 4*x**2)
ax[0, 2].set_title("0,2'deki plot")

ax[1, 0].plot(x, 2* x*2)
ax[1, 0].set_title("1,0'daki plot")

ax[1, 1].plot(x, 2* x**2)
ax[1, 1].set_title("1,1'deki plot")

ax[1, 2].plot(x, 3* x**2)
ax[1, 2].set_title("1,2'deki plot")

fig.suptitle("Bütün figür")
fig.show()
```

```python
fig, axs = plt.subplots(2,2)

x = np.linspace(0, 2* np.pi, 400)
y = np.sin(x)

axs[0,0].plot(x,y)
axs[0,0].label_outer()

axs[0,1].plot(x, y**2)
axs[0,1].label_outer()

axs[1,0].plot(x, -y)
axs[1,0].label_outer()

axs[1,1].plot(x, -y**2)
axs[1,1].label_outer()

```

```python
fig, axs = plt.subplots(2,2)

x = np.linspace(0, 2* np.pi, 400)
y = np.sin(x)

bar_x = ["C++", "C#", "Java", "Python"]
bar_y = [10, 30, 45, 80]

scatter_x = np.random.rand(50)
scatter_y = np.random.rand(50)
scatter_sizes =  (30*np.random.rand(50)) ** 2
scatter_colors = np.random.rand(50)

axs[0,0].plot(x,y)

axs[0,1].bar(bar_x, bar_y)

axs[1,0].scatter(scatter_x, scatter_y, s = scatter_sizes, c = scatter_colors, alpha = 0.5)

axs[1,1].pie(bar_y, labels=bar_x, autopct='%1.1f%%')

fig.show()
```

```python
# Color mesh
x = np.arange(300)
y = np.arange(300)
Z = np.random.rand(100,100)
Z2 = np.random.rand(100,100) * 2 + 1
Z3 = np.random.rand(100,100) * 3 + 2

Z4 = np.random.rand(100,100) * 4 + 3
Z5 = np.random.rand(100,100) * 5 + 4
Z6 = np.random.rand(100,100) * 6 + 5

Z7 = np.random.rand(100,100) * 7 + 6
Z8 = np.random.rand(100,100) * 8 + 7
Z9 = np.random.rand(100,100) * 9 + 8

Z_row1 = np.hstack([Z, Z2, Z3])
Z_row2 = np.hstack([Z4, Z5, Z6])
Z_row3 = np.hstack([Z7, Z8, Z9])

Z_total = np.vstack([Z_row1, Z_row2, Z_row3])
plt.pcolormesh(x,y,Z_total)

plt.colorbar()

plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.animation
import numpy as np

t = np.linspace(0,2*np.pi)
x = np.sin(t)

fig, ax = plt.subplots()
ax.axis([0,2*np.pi,-1,1])
l, = ax.plot([],[])

def animate(i):
    l.set_data(t[len(t)-i:], x[len(t)-i:])

ani = matplotlib.animation.FuncAnimation(fig, animate, frames=len(t))

from IPython.display import HTML
HTML(ani.to_jshtml())
```
