### For ve While Loopları (Döngüler)

Bir işi tekrar tekrar yapmak istediğimizde, döngü kullanıyoruz.

- Bir şeyi kaç kere tekrarlayacağımızı biliyorsak -> **FOR**
- Bir şeyi kaç kere tekrarlayacağımızı bilmiyorsak (Örn: kullanıcı -1 girene kadar) -> **WHILE**

- Bir listenin her bir elemanıyla bir işlem yapacaksak -> **FOR**
- Belli bir sonuç olana kadar tekrarlayacaksak -> **WHILE**




#### While 


Initilization

**while** CONDITION:

    - Statements
    - Update

```python
## Kullanıcı adı: mkemaltas
## Şifre: asdqwe123

# Kullanıcıdan k.adı ve şifre alıp, doğru girilene kadar tekrar tekrar soracağız.
# 3 kere yanlış girilirse hata versin.


username = input("Kullanıcı adı:")
password = input("Şifre: ")
trials = 1

while (username != "mkemaltas" or password != "asdqwe123") and trials < 3:
    
    if username != "mkemaltas" and password != "asdqwe123":
        print("Hatalı kullanıcı adı ve şifre! ({} hakkınız kaldı)".format(3 - trials)) 
    elif username != "mkemaltas":
        print("Hatalı kullanıcı adı! ({} hakkınız kaldı)".format(3 - trials)) 
    elif password != "asdqwe123":
        print("Hatalı şifre! ({} hakkınız kaldı)".format(3 - trials)) 
                
    username = input("Kullanıcı adı:")
    password = input("Şifre: ")
    
    trials += 1
      
if username == "mkemaltas" and password == "asdqwe123":
    print("Tebrikler, doğru kullanıcı adı ve şifre!")
else:
    print("Deneme haklarınız doldu.")
    
    
```

### For

%99 oranında listeleri gezmek amacıyla kullanıyoruz.

\["Mustafa", "Ahmet", "Gökçe", "Selin"\]


**for** _ELEMAN_ in _LISTE_:

    - statement
    
    
    
Bir işi kaç kere tekrarlayacağımızı bildiğimiz senaryolarda, "N kere şu işi yap" anlamında da kullanabiliyoruz:


**for** _ in range(N):

    - statement
    
#### range 

Başlangıç ve bitiş noktası belli olan bir sayı listesi oluşturuyor:

lst = range(x, y) dediğimizde, x'ten y'ye kadar bütün sayılardan oluşan bir liste oluşturuyor

lst = range(x,y,z) dediğimizde x'ten y'ye kadar z adımlar atlayarak bir liste oluşturuyor


**y'ler dahil değil!!**

```python
lst = range(10, 20)

print(list(lst))

lst = range(10, 20, 3)
print(list(lst))

lst = range(20, 10, -2)
print(list(lst))
```

```python
## Uzunlukları aynı olan isimlerden oluşan bir liste verildiğinde,
## her kelimenin birbirine denk gelen harflerinden yeni isimler oluşturan bir kod yazalım

isim_listesi = ["Arda", 
        "Mert", 
        "Onur", 
        "Seda", 
        "Ayşe", 
        "Utku", 
        "Gaye"]


## Harf sıraları üzerinde gezen ilk for loopum

for harf_sirasi in range(len(isim_listesi[0])):
    
    birlesen_kelime = ""
    # Her bir harf sırası için, bütün kelimelerin o sıradaki harflerini almam lazım.
    for isim in isim_listesi:
        birlesen_kelime += isim[harf_sirasi]
        
    print(birlesen_kelime)
```

```python
## Kullanıcıdan bir sayı alıp, satır ve sütun sayısı bu sayı olan bir kare çizin (* kullanarak)

# Örn:
# inp: 3

# ***
# ***
# ***


sayi = int(input("Bir sayı girin: "))

for _ in range(sayi): # her bir satırı temsil ediyor
    for _ in range(sayi): # her satırdaki bir yıldızı temsil ediyor
        print("*", end = "")
    print()
    

print()
    
for _ in range(sayi):
    print("*" * sayi)
```

## Object Oriented Programming (Class)

**class** CLASS_ISMI:

    def __init__(self, param1, param2): (constructor)
        self.param1 = param1

```python
class Pokemon:
    def __init__(self, name, _type, HP, power, trainer):
        self.name = name
        self.type = _type
        self.HP = HP
        self.power = power
        self.trainer = trainer
        
    def attack(self, attack_name):
        print("{} used {} ({} damage)".format(self.name, attack_name, self.power))

    def __str__(self):
        return "{}({})\nHP: {}\nAttack Power: {}\nTrainer: {}".format(self.name, self.type, self.HP, self.power, self.trainer)
    
pikachu = Pokemon("Pikachu", "Electric", 100, 40, "Ash Ketchum")
staryu = Pokemon ("Staryu", "Water", 120, 30, "Misty")
onix = Pokemon("Onix", "Rock", 500, 100, "Brock")


print(pikachu.trainer)

pikachu.attack("Thunderbolt")
staryu.attack("Water Gun")
onix.attack("Body Slam")

print("--------\n")
print(pikachu)
print("--------\n")
print(staryu)
print("--------\n")
print(onix)
```

```python
class Pokemon:
    def __init__(self, name, HP, attack, defense):
        self.name = name
        self.HP = HP
        self.attack = attack
        self.defense = defense
        
class FirePokemon(Pokemon):
    def __init__(self, name, HP, attack, defense):
        super().__init__(name, HP, attack, defense)
        self.type = "Fire"
        
    def speak(self):
        print("WRWAARGH ({})".format(self.type))
        
    def performAttack(self, target_pokemon):
        damage = self.attack - target_pokemon.defense
        
        if target_pokemon.type == "Grass":
            damage = damage * 2
            print("{} attacked {}, it's super effective! ({} damage)".format(self.name, target_pokemon.name, damage))
        else:
            print("{} attacked {}. ({} damage)".format(self.name, target_pokemon.name, damage))
    
        target_pokemon.HP -= damage
    def __str__(self):
        return "{}({})\nHP: {}\nAttack Power: {}, Defense: {}".format(self.name, self.type, self.HP, self.attack, self.defense) 
    
    
class GrassPokemon(Pokemon):
    def __init__(self, name, HP, attack, defense):
        super().__init__(name, HP, attack, defense)
        self.type = "Grass"
        
    def speak(self):
        print("NYGIIEGH ({})".format(self.type))
        
    def performAttack(self, target_pokemon):
        damage = self.attack - target_pokemon.defense
        
        if target_pokemon.type == "Water":
            damage = damage * 2
            print("{} attacked {}, it's super effective! ({} damage)".format(self.name, target_pokemon.name, damage))
        else:
            print("{} attacked {}. ({} damage)".format(self.name, target_pokemon.name, damage))
            
        target_pokemon.HP -= damage
        
    
    def __str__(self):
        return "{}({})\nHP: {}\nAttack Power: {}, Defense: {}".format(self.name, self.type, self.HP, self.attack, self.defense) 
    
class WaterPokemon(Pokemon):
    def __init__(self, name, HP, attack, defense):
        super().__init__(name, HP, attack, defense)
        self.type = "Water"
        
    def speak(self):
        print("GLGGULGLGU ({})".format(self.type))
        
    def performAttack(self, target_pokemon):
        damage = self.attack - target_pokemon.defense
        
        if target_pokemon.type == "Fire":
            damage = damage * 2
            print("{} attacked {}, it's super effective! ({} damage)".format(self.name, target_pokemon.name, damage))
        else:
            print("{} attacked {}. ({} damage)".format(self.name, target_pokemon.name, damage))
            
        target_pokemon.HP -= damage
        
    
    def __str__(self):
        return "{}({})\nHP: {}\nAttack Power: {}, Defense: {}".format(self.name, self.type, self.HP, self.attack, self.defense) 

charmender = FirePokemon("Charmender", 100, 30, 10)
squirtle = WaterPokemon("Squirtle", 100, 30, 20)
bulbasaur = GrassPokemon("Bulbasaur", 120, 20, 15)


print(charmender)
print("\n-------")
print(squirtle)
print("\n-------")
print(bulbasaur)

print("\n")
charmender.performAttack(bulbasaur)
print("\n-------")
print(bulbasaur)
```
