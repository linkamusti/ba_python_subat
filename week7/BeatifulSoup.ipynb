```python
from bs4 import BeautifulSoup


f = open("demo.html", "r")
html_body = f.read()

soup = BeautifulSoup(html_body, 'html.parser')
result = soup.prettify()
title = soup.title
body = soup.body

print(title.string)
#print(body)

result = soup.h1
print(result.string)

result = soup.h2
print(result.string)
print(result.name)

result = soup.find_all('h2')
print(result)
print("-"*50)
result = soup.find_all('div')
for div in result:
    print(div)
    print("-"*50)

ilk_div = result[0]
print(ilk_div.h2)
print(ilk_div.ul.li)
print(ilk_div.ul.find_all("li"))


print("-"*50)


result = soup.div.findChildren()
for child in result:
    print(child)
    print("-"*50)
    
    
    
print("-"*50)
result = soup.div.findNextSibling()
print(result)

print("-"*50)
result = soup.div.findNextSibling().findNextSibling()
print(result)
```
