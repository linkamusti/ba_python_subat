```python
import requests
from bs4 import BeautifulSoup



url = "https://www.n11.com/bilgisayar/dizustu-bilgisayar"
response = requests.get(url)


soup = BeautifulSoup(response.text, "html.parser")

urunler = []
fiyatlar = []

h3_list = soup.find_all("h3", {"class": "productName"})
for h3 in h3_list:
    if h3.string is not None:
        urunler.append(h3.string.strip())
        
        

proDetailDivs = soup.find_all("div", {"class": "proDetail"})
for div in proDetailDivs:
    fiyatlar.append(div.ins.text.replace("TL", "").strip())
    

for sayfa in range(2, 10):
    response = requests.get(url+"?pg="+str(sayfa))

    soup = BeautifulSoup(response.text, "html.parser")

    h3_list = soup.find_all("h3", {"class": "productName"})
    for h3 in h3_list:
        if h3.string is not None:
            urunler.append(h3.string.strip())



    proDetailDivs = soup.find_all("div", {"class": "proDetail"})
    for div in proDetailDivs:
        fiyatlar.append(div.ins.text.replace("TL", "").strip())
    

    
for i in range(len(urunler)):
    print(urunler[i], fiyatlar[i])
```
