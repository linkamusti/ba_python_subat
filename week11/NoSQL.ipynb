```python
#Mongo Insertion(one)

import pymongo
import ssl

myclient = pymongo.MongoClient("mongodb+srv://bapython:bapython@cluster0.p04bl.mongodb.net/bilgeadam?retryWrites=true&w=majority",ssl_cert_reqs=ssl.CERT_NONE)

mydb = myclient['BilgeAdam']
mycollection = mydb['products']

product = {
    "name": "Samsung S5",
    "price": 2000
}

result = mycollection.insert_one(product) # tek bir adet döküman eklemek için
print(result)
print(type(result))
print(result.inserted_id)
```

```python
#Mongo Insertion(many)

products = [
    {"name": "Samsung S6" , "price": 3000},
    {"name": "Samsung S7" , "price": 4000},
    {"name": "Samsung S8" , "price": 5000},
    {"name": "Samsung S9" , "price": 6000},
    {"name": "Samsung S10", "price": 7000},
    {"name": "Samsung S11", "price": 8000},
]

mycollection.insert_many(products)
```

```python
# Mongo Insertion(many)
# Inserting data with different structure (ikinciel fazladan eklendi)
products = [
    {"name": "Iphone 6" , "price": 7000, "ikinciel":True},
    {"name": "Iphone 7" , "price": 8000, "ikinciel":True},
    {"name": "Iphone XS" , "price": 9000, "ikinciel":True},
]

mycollection.insert_many(products)
```

```python
#Update

mycollection.update_one({'name': 'Samsung S8'},   
                        {'$set': {"name": "Samsung Galaxy S8"}}) # birinci parametre filtre, ikinci parametre değişiklik


mycollection.update_one({'name': 'Samsung S6'},   
                        {'$set': {"price": 1500}}) # birinci parametre filtre, ikinci parametre değişiklik

mycollection.update_many({'price' : { "$gt" : 5000}}, # $inc-> increment anlamına geliyo(arttırma)
                         {'$inc': {"price": 250}})
```

```python
#Delete 

mycollection.delete_one({'name': 'Samsung S5'})

mycollection.delete_many({'price': {'$lte': 5000 }}) # lte -> less than or equal to(küçükeşit)
```
