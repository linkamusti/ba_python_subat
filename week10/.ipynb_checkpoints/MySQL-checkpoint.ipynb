```python
#Create Table

mydb = connector.connect(
    host="db4free.net",
    user="bilgeadam",
    password="bilgeadam",
    database="bilgeadam_test")

mycursor = mydb.cursor() # create a cursor to use database

create_sql = "CREATE TABLE products(name VARCHAR(255), price FLOAT, description VARCHAR(255))"
mycursor.execute(create_sql)


mydb.close()
```

```python
#Insert new row (single row)

mydb = connector.connect(
    host="db4free.net",
    user="bilgeadam",
    password="bilgeadam",
    database="bilgeadam_test")

mycursor = mydb.cursor() # create a cursor to use database


while True:
    name = input("Name: ")
    price = input("Price: ")
    description = input("Description: ")

    insert_sql = "INSERT INTO products(name, price, description) VALUES(%s, %s, %s)"

    mycursor.execute(insert_sql, (name, price, description))
    
    try:
        mydb.commit() # değişiklikler database'e yansıtılıyor.
        print("{} tane kayıt eklendi.".format(mycursor.rowcount))
        print("son eklenen kaydın id numarası: {}".format(mycursor.lastrowid))
    except connector.Error as err:
        print("hata:", err)
    
    devam = input("devam etmek istiyor musunuz?(e/h)")
    if devam == "h":
        break
    
mydb.close()
```

```python
#Insert new rows (multiple rows)

mydb = connector.connect(
    host="db4free.net",
    user="bilgeadam",
    password="bilgeadam",
    database="bilgeadam_test")

mycursor = mydb.cursor() # create a cursor to use database


urun_listesi = []

while True:
    name = input("Name: ")
    price = input("Price: ")
    description = input("Description: ")
    
    urun_listesi.append((name, price, description))
    
    devam = input("devam etmek istiyor musunuz?(e/h)")
    if devam == "h":
        break
        

        
insert_sql = "INSERT INTO products(name, price, description) VALUES(%s, %s, %s)"

mycursor.executemany(insert_sql, urun_listesi) # listedeki her bir elemanı toptan ekler.

try:
    mydb.commit() # değişiklikler database'e yansıtılıyor.
    print("{} tane kayıt eklendi.".format(mycursor.rowcount))
    print("son eklenen kaydın id numarası: {}".format(mycursor.lastrowid))
except connector.Error as err:
        print("hata:", err)
    
mydb.close()
```
