```python
# HTML response

import requests

result = requests.get("https://www.imdb.com")
print(result.text)
```

```python
# API sample
# JSON response
import json

result = requests.get("https://jsonplaceholder.typicode.com/posts")

data = json.loads(result.text)
print(data[0]['title'])
```

```python
from IPython.core.display import Image, display

pokemon_url = "https://pokeapi.co/api/v2/pokemon?limit=500"

result = requests.get(pokemon_url)
data = json.loads(result.text)
data = data['results']
    

for poke in data:
    print(poke)

pokemon_name = input("Bir pokemon ismi girin: ")

for pokemon in data:
    if pokemon['name'] == pokemon_name:
        result = requests.get(pokemon['url'])
        pokemon_data = json.loads(result.text)
        sprites = pokemon_data['sprites']
        display(Image(url = sprites['back_default'], width=100, height=100))
        display(Image(url = sprites['front_default'], width=100, height=100))        
```

```python
from IPython.core.display import Image, display
import random

pokemon_url = "https://pokeapi.co/api/v2/pokemon?limit=1500"

result = requests.get(pokemon_url)
data = json.loads(result.text)
data = data['results']
    
pokemon = random.choice(data)

result = requests.get(pokemon['url'])
pokemon_data = json.loads(result.text)
sprites = pokemon_data['sprites']
print(pokemon_data['name'])
display(Image(url = sprites['back_default'], width=100, height=100))
display(Image(url = sprites['front_default'], width=100, height=100))        
```
